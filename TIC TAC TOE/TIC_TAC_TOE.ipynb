```python
board=["-","-","-","-","-","-","-","-","-"]
count=0
 
 
status=1
current_player="X"
 
def board_info():
    print(board[0],"|",board[1],"|",board[2])
    print(board[3],"|",board[4],"|",board[5])
    print(board[6],"|",board[7],"|",board[8])
    
def get_inp():
    
    i=int(input("ENter the position form 1-9"))
    board[i-1]=current_player
    board_info()
    print("\n\n")
    
 
 
    
    
def win_status():
    a=rows()
    b=columns()
    c=diagonals()
    if(a!=None):
        print("\n\n",a,"Wins")
    if(b!=None):
        print("\n\n",b,"Wins")
    if(c!=None):
        print("\n\n",c,"Wins")
  
        
 
    
def rows():
    global status
    row1= board[0]==board[1]==board[2] !="-"
    row2= board[3]==board[4]==board[5] !="-"
    row3= board[6]==board[7]==board[8] !="-"
    if(row1 or row2 or row3):
        status=0
    if(row1):
        return(board[0])
    if(row2):
        return(board[3])
    if(row3):
        return(board[6])
 
    
def columns():
    global status
    col1= board[0]==board[3]==board[6] !="-"
    col2= board[1]==board[4]==board[7] !="-"
    col3= board[2]==board[5]==board[8] !="-"
    
    if(col1 or col2 or col3):
        status=0
    if(col1):
        return(board[0])
    if(col2):
        return(board[1])
    if(col3):
        return(board[2])
    
    
def diagonals():
    global status
    
    d1=board[0]==board[4]==board[8]!="-"
    d2=board[2]==board[4]==board[6]!="-"
    
    if(d1 or d2):
        status=0
    if(d1):
        return board[0]
    if(d2):
        return board[2]
    
def flip():
    global current_player
    
    
    if(current_player=="X"):
        current_player="Y"
    elif(current_player=="Y"):
         current_player="X"
        
    
    
    
    
board_info()    
while status:
    
    print("\n\n")
    get_inp()
    win_status()
    flip()
    count=count+1
    if(count==9 and win_status!="X" and win_status!="Y"):
        print("\n\n\n\nSorry Bruh Tie")
        break
```
